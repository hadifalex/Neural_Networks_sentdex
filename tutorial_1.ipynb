[Link to the video](https://www.youtube.com/watch?v=Wo5dMEP_BbI).

The first column demarked as `x` is the input which contains N components.
The first `neuron` is the the first circle at the second column.
This is is represented by the sum of all inputs each with its own corresponding weight.



The very first element of the INPUT is $x_{1,1}$. The first neuron element is then given by

$z_{1,1} = w_{1,1,1}*x_{1,1} + w_{1,2,1}*x_{1,2} +...$

$z_{1,1} = \sum{w_{1,i,1}*x_{1,i}} + b_{1,1}$

The first LAYER is the first column, with each NEURON having its own weightings.
Since in this scenario we have only 4 circles, the first layer has 4 neurons:

$z_1 = \sum{w_{1,i,1}*x_{1,i}} + b_{1,1} +\sum{w_{1,i,2}*x_{1,i}}+ b_{1,2} + \sum{w_{1,i,3}*x_{1,i} + b_{1,3} +\sum{w_{1,i,4}*x_{1,i}}+ b_{1,4}}$

In short, each column (layer) here can be written as 

$z_1=\sum_{i,j}{w_{1,i,j}*x_{1,i}}+ b_{1,j}$

Where $i$ the number of elements in the input, $j$ the number of weightings within the first neuron, and  $b_j$ is just a bias that we are adding to each neuronal component.

After this sum is done, we need to run each neuron through an activation function.
I am not certain what that is or does, but the expression is 

$y_1 = max(0,z_1)$

The above expression represents the OUTPUT from the first LAYER!

This is now repeated for the second layer which takes $y_1$ as an INPUT.

You then take the average, and calculate what is known as the LOSS FUNCTION (the wrong the neural network is).



The output layer in this case has two possibilities (failure neuron vs non-failure neuron, cat vs dog, etc...)

The idea is that if you train the weightings, the bottom neuron will fire if it is a cat and the top neuron if it is a dog.

Every neuron is connected to every neuron of the next weight with its own unique weight. And every neuron has its own bias. This means that some neurons may end up being MORE IMPORTANT that others.

In the following case, if you have 60 neurons in total, you also have 60 neurons.
With this configuration, it also means that you have 704 weightings to deal with. This means that there are a total of 764 parameters taht can be tuned to what we need.





> The difficulty of machine learning and deep learning is NOT the construction of the network but the correct way to TUNE it.

```python
# Creating the first neuron
## every neuron is connected to every single neuron from the previous layer.

inputs  = [1.2,5.1,2.1]     # Let's assume that these are the outputs from the previous layer.
weights = [3.1,2.1,8.7]     # the weights that are associated with the neurons
bias    = 3                 # every layer has it's own bias

output  = inputs[0]*weights[0] + inputs[1]*weights[1]+inputs[2]*weights[2] + bias # this is the output from our neural network
output
```
